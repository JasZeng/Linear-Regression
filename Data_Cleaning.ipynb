Write function to change from #M,#B,#T, #k into numbers

```python
def turn_big_numbers(large_num):
    try:
        return float(large_num)
    except: 
        try:unit=large_num[-1]
        except: return ('no -1')
        multiplier=0
        if unit=='T':
            multiplier=1000000000000
        elif unit=='B':
            multiplier=1000000000
        elif unit=='M':
            multiplier=1000000
        elif unit=='k':
            multiplier=1000
        num=float(large_num[0:(len(large_num)-1)])
        return num*multiplier
```

Now changing string with % into float

```python
def turn_perc_str(perc_str):
    try: 
        return (float(perc_str.strip('%')) / 100.0)
    except:
        return perc_str
```

```python
change_list=['Profit_margin','Operating_margin','Return_on_asset','Return_on_equity','Quarterly_revenue_growth']
for col in change_list:
    nasdaq_df[col]=nasdaq_df[col].apply(turn_perc_str)
```

Now changing '[price]' into float

```python
def del_brckt(price):
    try: 
        return float(price[1:(len(price)-1)])
    except:
        return float('NaN')
```

```python
change_list=['current_price','last_quarter_price','last_year_price']
for col in change_list:
    nasdaq_df[col]=nasdaq_df[col].apply(del_brckt)
```
